# Shuffling the Dataset

```python
iris=iris.sample(frac=1)
```

```python
iris.head(20)
```

# Encoding the Target Data

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
```

```python
label_encode = LabelEncoder()
```

```python
iris["Encode_Species"] = label_encode.fit_transform(iris["Species"])
```

```python
iris.head(30)
```

# Converting label column into an array

```python
en_iris = label_encode.fit_transform(iris["Species"])
en_iris
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = iris.iloc[:,:-2].values
```

```python
x
```

```python
x.shape
```

```python
y = en_iris
y
```

# Train_Test_Split

```python
x_train,x_test,y_train,y_test = train_test_split(x, y, test_size = 0.3, random_state = 100)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
decisiontree = DecisionTreeClassifier()
```

```python
decisiontree.fit(x_train, y_train)
```

```python
predict = decisiontree.predict(x_test)
```

# Testing accuracy score

```python
from sklearn.metrics import accuracy_score
```

```python
acc_test = accuracy_score(y_test, predict)
```

```python
acc_test
```

# Importing Confusion_matrix , Classification_report

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
print(classification_report(y_test, predict))
```

```python
cmm_test = confusion_matrix(y_test, predict)
cmm_test
```

```python
cmm=pd.DataFrame(data=cmm_test, index=['setosa','versicolor','virginica'], columns=['setosa','versicolor','virginica'])
plt.figure(figsize=(10,7))

sns.heatmap(cmm,annot=True,cmap='Blues')
```

# Testing on a Random Test Case

```python
test = np.array([(5.2,3.5,1.5,0.2)])
```

```python
predict_test = decisiontree.predict(test)
```
