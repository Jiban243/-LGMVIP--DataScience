# Train_Test_Split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train ,X_test, Y_train, Y_test = train_test_split(x, y, test_size = 0.2)
```

```python
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

# Applying KNN

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 4)
knn.fit(X_train, Y_train)
```

```python
predict_test = knn.predict(X_test)
predict_test
```

```python
predict_train=knn.predict(X_train)
```

```python
from sklearn.metrics import accuracy_score
```

```python
acc_test = accuracy_score(Y_test, predict_test)
```

```python
acc_test
```

```python
from sklearn.metrics import confusion_matrix
cm_test = confusion_matrix(Y_test, predict_test)
cm_test
```

```python
cm_train = confusion_matrix(Y_train, predict_train)
cm_train
```

```python
l_train=[]
l_test=[]
for i in range(1,20):
    knn=KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train,Y_train)
    x=confusion_matrix(Y_train,knn.predict(X_train))
    y=confusion_matrix(Y_test,knn.predict(X_test))
    l_train.append((x[0][1]+x[0][2]+x[1][2]+x[1][0]+x[2][0]+x[2][1])/x.sum())
    l_test.append((y[0][1]+y[0][2]+y[1][2]+y[1][0]+y[2][0]+y[2][1])/y.sum())
    

    
    
    
    
```

```python
X_cvtrain
```

```python
Y_cvtest
```

```python
l_cvtrain=[]
l_cvtest=[]
for i in range(1,20):
    knn=KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_cvtrain,Y_cvtrain)
    x=confusion_matrix(Y_cvtrain,knn.predict(X_cvtrain))
    y=confusion_matrix(Y_cvtest,knn.predict(X_cvtest))
    l_cvtrain.append((x[0][1]+x[0][2]+x[1][2]+x[1][0]+x[2][0]+x[2][1])/x.sum())
    l_cvtest.append((y[0][1]+y[0][2]+y[1][2]+y[1][0]+y[2][0]+y[2][1])/y.sum())
    
```

```python
plt.plot(range(1,20),l_cvtrain,label="train error")
plt.plot(range(1,20),l_cvtest,label="test error")
plt.xlabel("k values")
plt.ylabel("error")
plt.legend()
```
