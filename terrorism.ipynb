# Importing Libraries

```python
# Importing the Header Files
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import warnings
```

```python
terrorism=pd.read_csv('terrorism.csv',encoding='latin-1')
```

# Renaming the columns

```python
terrorism.rename(columns={'iyear':'Year','imonth':'Month','extended':'Extended','iday':'Day','country_txt':'Country',
           'provstate':'State','region_txt':'Region','attacktype1_txt':'AttackType','target1':'Target',
           'nkill':'Killed','nwound':'Wounded','summary':'Summary','gname':'Group','targtype1_txt':'Target_type',
           'weaptype1_txt':'Weapon_type','motive':'Motive','city':'City','latitude':'Latitude','longitude':'Longitude'},inplace=True)
```

# Deleting the unnecassary columns

```python
terrorism=terrorism[['Year','Month','Extended','Day','Country','State','Region','City','Latitude','Longitude',
           'AttackType','Killed','Wounded','Target','Summary','Group','Target_type','Weapon_type','Motive']]
```

```python
terrorism.head(10)
```

```python
terrorism.describe
```

# Checking for null values

```python
terrorism.isnull().sum()
```

```python
val=terrorism['Country'].value_counts().head(10)
val
```

# Most affected countries

```python
plt.figure(figsize = (10,5))
sns.barplot(terrorism['Country'].value_counts()[:15].index,terrorism['Country'].value_counts()[:15].values)
plt.title('Top 10 Countries Affected')
plt.xlabel('Countries')
plt.ylabel('count')
plt.xticks(rotation=90)
plt.show()
```

 Most affected country - Iraq
